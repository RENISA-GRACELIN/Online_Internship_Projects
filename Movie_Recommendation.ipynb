```python
df_features = df[['Movie_Genre', 'Movie_Keywords', 'Movie_Tagline', 'Movie_Cast', 'Movie_Director']].fillna('')
x = (
    df_features['Movie_Genre'] + ' ' +
    df_features['Movie_Keywords'] + ' ' +
    df_features['Movie_Tagline'] + ' ' +
    df_features['Movie_Cast'] + ' ' +
    df_features['Movie_Director']
)
```

# **Get Feature Text Conversion to Token**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

# **Get Similarity score using Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_score=cosine_similarity(x)
```

```python
Similarity_score
```

```python
Similarity_score.shape
```

# **Get Movie Name as Input from user and Validate for closet spelling**

```python
Fav_Movie=input('Enter your favourite movie name: ')
```

```python
All_movies_title_list=df['Movie_Title'].tolist()
```

```python
import difflib
```

```python
Movie_Recommendation=difflib.get_close_matches(Fav_Movie,All_movies_title_list)
```

```python
print(Movie_Recommendation)
```

```python
close_match=Movie_Recommendation[0]
print(close_match)
```

```python
Index_of_close_match_movie=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
print(Index_of_close_match_movie)
```

```python
Recomm_score=list(enumerate(Similarity_score[Index_of_close_match_movie]))
print(Recomm_score)
```

```python
len(Recomm_score)
```

# **Get all the movies sort based on Recommendation Score with respect to Favourite Movie**

```python
Sorted_similar_movies=sorted(Recomm_score,key=lambda x:x[1],reverse=True)
print(Sorted_similar_movies)
```

```python
print('Top 30 Movies suggested for you:\n')
i=1
for movie in Sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['Movie_Title'].values[0]
  if (i<31):
    print(i,'.',title_from_index)
    i+=1
```

# **Top 10 Movie Recommendation System**

```python
Movie_name=input('Enter your favourite movie name:')
list_of_all_titles=df['Movie_Title'].tolist()
Find_close_match=difflib.get_close_matches(Movie_name,list_of_all_titles)
close_match=Find_close_match[0]
Index_of_Movie=df[df.Movie_Title==close_match]['Movie_ID'].values[0]
Recomm_score=list(enumerate(Similarity_score[Index_of_Movie]))
Sorted_similar_movies=sorted(Recomm_score,key=lambda x:x[1],reverse=True)
print('Top 10 Movies suggested for you:\n')
i=1
for movie in Sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.Movie_ID==index]['Movie_Title'].values
  if (i<11):
    print(i,'.',title_from_index)
    i+=1
```
